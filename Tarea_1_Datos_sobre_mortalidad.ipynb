## 1. Descargando los datos

Comenzamos cargando las bibliotecas necesarias.

```python
import os  # Para manejo de archivos y directorios
import urllib.request # Una forma estandard de descargar datos
# import requests # Otra forma no de las librerías de uso comun

import datetime # Fecha de descarga
import pandas as pd # Solo para ver el archivo descargado
import zipfile # Descompresión de archivos
```

Creamos los subdirectorios para guardar los datos de manera ordenada.

```python
# pwd
print(os.getcwd())

#  Estos son los datos que vamos a descargar y donde vamos a guardarlos
mortalidad_2020_url = "https://www.inegi.org.mx/contenidos/programas/mortalidad/datosabiertos/defunciones/2020/conjunto_de_datos_defunciones_registradas_2020_csv.zip"
mortalidad_2020_archivo = "mortalidad_2020"

salud_particulares_2020_url = "https://www.inegi.org.mx/contenidos/programas/salud/datosabiertos/conjunto_de_datos_esep_2020_csv.zip"
salud_particulares_2020_archivo = "salud_2020.zip"
subdir = "./data/" 
```

Revisamos si ya existen los subdirectorios. Si no es así los creamos. 

Enseguida, abrimos las url de los archivos, extraemos contenidos de los zip y los guardamos.

```python
# Revisamos si el archivo existe. Si no existe, se procede a crearlo.
if not os.path.exists(mortalidad_2020_archivo):
    # Revisamos si el directorio existe. Si no es así procedemos a crearlo.
    if not os.path.exists(subdir):
        #Creamos el directorio 
        os.makedirs(subdir)
    # Abrimos la url y guardamos el objeto en mortalidad_2020_archivo
    urllib.request.urlretrieve(mortalidad_2020_url, subdir + mortalidad_2020_archivo)  
    # Extraemos el contenido del zip y lo guardamos.
    with zipfile.ZipFile(subdir + mortalidad_2020_archivo, "r") as zip_ref:
        zip_ref.extractall(subdir)

if not os.path.exists(salud_particulares_2020_archivo):
    urllib.request.urlretrieve(salud_particulares_2020_url, subdir + salud_particulares_2020_archivo) 
    with zipfile.ZipFile(subdir + salud_particulares_2020_archivo, "r") as zip_ref:
        zip_ref.extractall(subdir)
```

Creamos un archivo tipo txt, con la información de las bases de datos y las fuentes de donde obtuvimos la información.

```python
with open(subdir + "info.txt", 'w') as f:
  f.write("Archivos sobre mortalidad \n")
  info = """
  mortalidad_2020: La estadística de defunciones registra la información sobre el 
  fenómeno de la mortalidad en el país,que permite conocer y comparar el volumen, 
  tendencia y características de la mortalidad en los diferentes ámbitos geográficos
  del país, conformando un insumo para el análisis y evaluación de acciones dirigidas 
  a la elaboración de programas de salud pública.

  En esta base de datos Se captan las defunciones registradas en todo el país en el 
  Sistema de Registro Civil, así como las defunciones accidentales y violentas en 
  las que intervienen el Servicio Médico Forense y el Ministerio Público.

  salud_2020: Estas estadísticas abordan los datos sobre servicios, morbilidad 
  hospitalaria, así como recursos humanos y materiales proporcionados por las 
  instituciones privadas.


  Los datos se obtuvieron de datos abiertos del INEGI con fechas de 27 de agosto 2022.

  """ 
  f.write(info + '\n')
  f.write("Descargado el " + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") + "\n")
  f.write("Desde: " + mortalidad_2020_url + "\n")
  f.write("Nombre: " + mortalidad_2020_archivo + "\n")
  f.write("Agregados salud en establecimientos particulades descargados desde: " + salud_particulares_2020_url + "\n")
  f.write("Nombre: " + salud_particulares_2020_archivo + "\n")
```
